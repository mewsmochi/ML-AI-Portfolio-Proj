```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#importing the video games csv file
video_games = pd.read_csv('video_game_sales.csv')
```

##### Dataset provided by *The Devastator* on Kaggle

```python
#finding the total number of game genres in the dataset
genre_counts = video_games.Genre.value_counts()
print(genre_counts)
```

#### There are 12 genres: Sports, Action, Shooter, Platform, Racing, Role-Playing, Figting, Adventure, Simulation, Puzzle, Strategy, and Miscellaneous

#### The five most recurring genres are Sports, Action, Shooter, Platform, and Racing.

```python
#Separating the rows of the dataset by game genre
all_sports_games = video_games[(video_games.Genre == 'Sports')]
all_action_games = video_games[(video_games.Genre == 'Action')]
all_shooter_games = video_games[(video_games.Genre == 'Shooter')]
all_platform_games = video_games[(video_games.Genre == 'Platform')]
all_racing_games = video_games[(video_games.Genre == 'Racing')]
```

#### The function "sum_of_column" returns the sum of numbers in a specified column.

```python
#function to find the sum of numbers in a column
def sum_of_column(df, column):
    sum = df[column].sum(axis=0)
    return sum
```

```python
# finding the sum of units sold globally (in millions) of each gaming genre
sports_sum = sum_of_column(all_sports_games, 'Global')
action_sum = sum_of_column(all_action_games, 'Global')
shooter_sum = sum_of_column(all_shooter_games, 'Global')
platform_sum = sum_of_column(all_platform_games, 'Global')
racing_sum = sum_of_column(all_racing_games, 'Global')
```

```python
#Two lists, one of the top 5 video game genres, and the other the amount of units sold globally in millions 
genres = ['Sports', 'Action', 'Shooter', 'Platform', 'Racing']
units_sold = [sports_sum, action_sum, shooter_sum, platform_sum, racing_sum]
```

```python
plt.bar(genres, units_sold, color = 'pink', 
        width = 0.4)
 
plt.xlabel("Genres")
plt.ylabel("Units Sold (in millions)")
plt.title("Top 5 Genres & Units Sold Globally")
plt.show()
```

#### The genre that sold the most games is the Sports genre, followed by Action, Platform, Shooter, and Racing. However, the amount of units sold aren't spread equally throughout the games in the genre, but are rather usually dominated by a specific game series or publisher that is dominating the genre. For example, the sports genre:

```python
#putting all the publishers in the sports game genre in a list
sports_publishers = all_sports_games['Publisher'].tolist()
#putting all the units sold globally (in millions) in the sports game genre in a list
sports_global_sales = all_sports_games['Global'].tolist()

#Changing all the values in sports_global_sales to an integer
sports_global_sales_int = [int(item) for item in sports_global_sales]
#Changing all the values in sports_publishers to a string
sports_publishers_str = [str(item) for item in sports_publishers]

#Creating a bar chart of all Game Publishers in the Sports Genre and their number of units sold
plt.bar(sports_publishers_str, sports_global_sales_int, color = 'purple', width = 1)
plt.xlabel("Sports Genre Publishers")
plt.xticks(rotation=90)
plt.ylabel("Units Sold (in millions)")
plt.title("Game Publishers in the Sports Genre and their Units Sold")
plt.show()
```

#### As shown in the graph, the number of units sold are nowhere near distributed evenly. Nintendo has completely dominated this genre, providing more proof to the argument that it isn't the genres that are super popular, but actually just huge gaming publishers influencing the numbers.
